```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df_1 = pd.read_csv('Data/superhero_info.csv')
df_1.head()
```

```python
df_2 = pd.read_csv("Data/superhero_powers - superhero_powers.csv")
df_2.head()
```

```python
## adding expand=True
df_1['Hero|Publisher'].str.split('|',expand=True)
## Save the 2 new columns into the dataframe
df_1[['Hero','Publisher']] = df_1['Hero|Publisher'].str.split('|', expand=True)
## drop the original column 
df_1 = df_1.drop(columns=['Hero|Publisher'])
# Display the first 5 rows of the dataframe
df_1.head()
```

```python
#Moving the Hero column to the front 
# Pop Name column
first_column = df_1.pop('Hero')
# Insert Name column as the first column
df_1.insert(0, 'Hero', first_column)
# Now we see the Hero column was moved to the front!
df_1.head()
```

```python
#Creating a sample value to convert same from string to dictionary
test_msm = df_1.loc[0, 'Measurements']
# Display the sample value's datatype and the sample value
print(type(test_msm))
test_msm
```

```python
# Replace single quotation with the double quotation marks
test_msm = test_msm.replace("'",'"')
test_msm
```

```python
# Convert string to dictionary
test_msm = json.loads(test_msm)
print(type(test_msm))
test_msm
```

```python
#Converting entire column from string to dictionary
# Use .str.replace to replace all single quotes
df_1['Measurements'] = df_1['Measurements'].str.replace("'",'"')
# Apply the json.loads to the full column
df_1['Measurements'] = df_1['Measurements'].apply(json.loads)
df_1['Measurements'].head()
```

```python
# Check a sample value after transformation
test_msm = df_1.loc[0, 'Measurements']
# Display the sample's datatype and the sample
print(type(test_msm))
test_msm
```

```python
#Unpack measurements columns of dictionaries into separate columns
# Unpack height_weight into a pandas series
height_weight = df_1['Measurements'].apply(pd.Series)
# Display the pandas series
height_weight.head()
```

```python
# combine Height and Weight columns to the dataframe
df_1 = pd.concat((df_1, height_weight), axis = 1)
df_1.head()
```

```python
# Drop Measurements column which has been replaced by Height and Weight
df_1 = df_1.drop(columns=['Measurements'])
```

```python
# Confirm Measurements column has been removed
df_1.head()
```

```python
# Remove spaces and unit of measures in Height column
to_replace = [' ', 'cm']
for char in to_replace:
    df_1['Height'] = df_1['Height'].str.replace(char, '', regex=False)
df_1['Height'].head()
```

```python
# Convert weight column to an integer
df_1["Height"] =pd.to_numeric(df_1["Height"]).astype(int)
```

```python
# Check datatype of the height column
df_1["Height"].dtype
```

```python
# Remove spaces and unit of measure to convert Weight column into numeric 
to_replace = [' ', 'kg']
for char in to_replace:
    df_1['Weight'] = df_1['Weight'].str.replace(char, '', regex=False)
df_1['Weight'].head()
```

```python
df_1["Weight"] =pd.to_numeric(df_1["Weight"]).astype(int)
df_1["Weight"].dtype
```

```python
#One-hot encoding the superpowers 
# Rename column to align with df_1 column name
df_2 = df_2.rename(columns = {"hero_names": "Hero"})
df_2.head()
```

```python
#List of powers 
df_2.loc[0,'Powers']
```

```python
# Display the data type for the sample of Powers
print(type(df_2.loc[0,'Powers']))
```

```python
# Convert to series 
powers_series = df_2["Powers"].str.split(",")
powers_series
```

```python
powers_series.explode("Powers")
```

```python
super_powers = powers_series.explode("Powers").dropna().sort_values().unique()
super_powers
```

```python
# Loop for one-hot-encoding 
for col in super_powers:
    df_2[col] = df_2['Powers'].str.contains(col)
```

```python
df_2.head()
```

```python
# Drop Powers column
df_2 = df_2.drop(columns=['Powers'])
```

```python
df_2.head()
```

```python
# Combine df_1 and df_2
df = pd.merge(df_1, df_2, on='Hero')
df.head()
```

```python
#Average weight of having super speed vs. not 
df.groupby('Super Speed')['Weight'].mean()
```

```python
#Average height for each publisher 
df.groupby('Publisher')['Height'].mean()
```
